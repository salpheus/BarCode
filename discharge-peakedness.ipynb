#### Calculate peakedness from max record

```python
qpeakCON = np.nanmax(congo[' Original'])/np.nanmean(congo[' Original'])
qpeakAMG = np.nanmax(amguema[' Original'])/np.nanmean(amguema[' Original'])
qpeakJAM = np.nanmax(jamuna[' Original'])/np.nanmean(jamuna[' Original'])
qpeakWRA = np.nanmax(wrangell[' Original'])/np.nanmean(wrangell[' Original'])
qpeakWAI = np.nanmax(waitaki[' Original'])/np.nanmean(waitaki[' Original'])
qpeakKOS = np.nanmax(koshi[' Original'])/np.nanmean(koshi[' Original'])
qpeakSSK = np.nanmax(southsaskatchewan[' Original'])/np.nanmean(southsaskatchewan[' Original'])
qpeakTAN = np.nanmax(tanana[' Original'])/np.nanmean(tanana[' Original'])
qpeakIRA = np.nanmax(irrawaddy[' Original'])/np.nanmean(irrawaddy[' Original'])
qpeakAMU = np.nanmax(amudarya[' Original'])/np.nanmean(amudarya[' Original'])
qpeakRAK = np.nanmax(rakaia[' Original'])/np.nanmean(rakaia[' Original'])
```

```python
## Calculate CV of peak discharge
congoCV = stats.variation(con_maxannual[' Original'][~np.isnan(con_maxannual[' Original'])])
irrawaddyCV = stats.variation(ira_maxannual[' Original'][~np.isnan(ira_maxannual[' Original'])])
jamunaCV = stats.variation(jam_maxannual[' Original'][~np.isnan(jam_maxannual[' Original'])])
amudaryaCV = stats.variation(amd_maxannual[' Original'][~np.isnan(amd_maxannual[' Original'])])
rakaiaCV = stats.variation(rai_maxannual[' Original'][~np.isnan(rai_maxannual[' Original'])])
waitakiCV = stats.variation(wai_maxannual[' Original'][~np.isnan(wai_maxannual[' Original'])])
koshiCV = stats.variation(kos_maxannual[' Original'][~np.isnan(kos_maxannual[' Original'])])
southskCV = stats.variation(ssk_maxannual[' Original'][~np.isnan(ssk_maxannual[' Original'])])
tananaCV = stats.variation(tan_maxannual[' Original'][~np.isnan(tan_maxannual[' Original'])])
wrangellCV = stats.variation(wra_maxannual[' Original'][~np.isnan(wra_maxannual[' Original'])])
amguemaCV = stats.variation(amg_maxannual[' Original'][~np.isnan(amg_maxannual[' Original'])])
```
